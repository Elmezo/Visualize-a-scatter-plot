```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd 
import numpy as np
from sklearn_extra.cluster import KMedoids
marker_colors = np.arange(start=0, stop=150, step=1)
```

```python
iris = datasets.load_iris()
iris_df=pd.DataFrame(iris.data)
iris_df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid']
iris_df.dropna(how="all", inplace=True)

iris_X=iris_df.iloc[:,[0,1,2,3]]
```

```python
iris_X
```

```python
plt.scatter(iris_X['sepal_len'], iris_X['sepal_wid'],c=marker_colors, cmap='viridis')
plt.title("petal width vs petal length")
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

```python
plt.scatter(iris_X['petal_len'], iris_X['petal_wid'],c=marker_colors, cmap='viridis')
plt.title("petal width vs petal length")
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

```python
kmedoids = KMedoids(n_clusters=2,metric="cosine",init="random" ,random_state=0)
kmedoids.fit_predict(iris_X)
plt.scatter(iris_X['sepal_len'], iris_X['sepal_wid'],c=kmedoids.labels_,cmap='rainbow')
```

```python
kmedoids = KMedoids(n_clusters=2,metric="cosine",init="random" ,random_state=0)
kmedoids.fit_predict(iris_X)
plt.scatter(iris_X['petal_len'], iris_X['petal_wid'],c=kmedoids.labels_,cmap='rainbow')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
def f(x, y):
    return np.cos(x*y*100)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 90)
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('3D data');
```

## saving the figures in .png images

```python
fig.savefig('image1.png', bbox_inches='tight')
```

```python
fig2 = plt.figure(figsize=(10,6))
ax2=plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

# saving the figures in .png images

```python
fig2.savefig('image2.png', bbox_inches='tight')
```

```python
fig3 = plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
```

```python
ax.view_init(60, 35)
fig3
```

# saving the figures in .png images

```python
fig3.savefig('image3.png', bbox_inches='tight')
```

```python
fig4 = plt.figure(figsize=(20,10))
plt.contour(X, Y, Z, colors='black');
plt.colorbar();
```

# saving the figures in .png images

```python
fig4.savefig('image4.png', bbox_inches='tight')
```
